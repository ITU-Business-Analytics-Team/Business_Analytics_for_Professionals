# **Time Series Analysis**
## Intermittent Time Series

## Problem Statement

For some products such as spare parts demand is not always occurs. In some periods there will be zero demand which is not easy to forecast using traditional time series analysis. 

```python
!pip install sktime
```

```python
import numpy as np
from sktime.datasets import load_PBS_dataset
from sktime.forecasting.croston import Croston
```

The Pharmaceutical Benefits Scheme (PBS) is  consists of the number of scripts for drugs sold each month. Most of the months there are no demands (Data for "Forecasting: Principles and Practice" (3rd Edition)) 




```python
pbs = load_PBS_dataset()
pbs.plot()
```

Split the data into training and test dataset

```python
pbs_train = pbs[-90:-66]
pbs_test = pbs[-66:-60]
pbs_train.plot()
pbs_test.plot()
```

```python
import matplotlib.pyplot as plt

mdl = Croston(smoothing=0.3)
mdl.fit(pbs_train)

ypred = mdl.predict(fh = np.arange(1,7))

pbs_train.plot(label = "Train")
pbs_test.plot(label = "Test")
ypred.plot(label = "Forecasted")
plt.grid()
plt.legend()
```
